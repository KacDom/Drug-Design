```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
```

```python
smiles = Chem.SmilesMolSupplier('AMCase.smi')
print(smiles)
```

```python
scaffoldList = []
for i in smiles:
    scaffoldList.append(i)
Draw.MolsToGridImage(scaffoldList, molsPerRow = 6)
```

```python
from rdkit.Chem.Scaffolds import MurckoScaffold
from rdkit.Chem.Scaffolds.MurckoScaffold import MurckoScaffoldSmiles
from rdkit.Chem import MolFromSmiles
```

```python
with open('scaffolds.txt', 'r') as scaffolds:
    lista = []
    lista1 = []
    
    for i in scaffolds:
        x = MurckoScaffoldSmiles(i)
        lista1.append(x)
        y = MolFromSmiles(x)
        lista.append(y)
    smiles_dict = {el: lista1.count(el) for el in set(lista1)}
smiles_dict
    
```

```python
porownanie = list(zip(scaffoldList,lista))
zipLista = []
for scc1,scc2 in porownanie:
    zipLista += scc1,scc2
```

```python
Draw.MolsToGridImage(zipLista, molsPerRow = 2)
```
